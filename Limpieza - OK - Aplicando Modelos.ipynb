# Aplicacion Modelos

```python
# Creamos el arreglo X sin la columna supervivientes, ya que seran nuestro datos de entrada
X = np.array(train.drop(['Survived'], 1))
```

```python
# Creamos el arreglo Y solo con columna supervivientes, ya que seran nuestro datos de salida
y = np.array(train['Survived'])
```

A continuacion realizamos el split test, para tener los datos de entrenamiento y prueba. Recordemos que como test_size usamos
0.2 ya que significa que usamos solo en 20% de los datos para las pruebas. Pero esto puede variar segun nuestra cantidad de
datos que tengamos.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

#  Regresion Logistica

Ahora como primer modelo, la regresion logistica, pero porque no la lineal, bueno la lineal solo nos sirve si tenemos una variable
Continua, mientra que el de regresion logistica nos sirve cuando nuestro datos de interes es dicotomico, osea toma los valores
si/no o 1/0 como la columna ['Survived']

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier


relog = LogisticRegression()
#Entrenamos el modelo con el comando FIT
relog.fit(X_train, y_train)
# Realizamos las predicciones para probar nuestro modelo con el comando predict
Y_pred_relog = relog.predict(X_test)
print(f"PrecisiÃ³n de la predicciÃ³n %.2f" %(relog.score(X_train, y_train)*100),"%")
```

Como vemos el modelo nos dio una prediccion del 80.03% que no esta mal para el primer modelo. Ahora veremos que sucederia con
los datos normalizados

#  k-Nearest Neighbor.

Antes de comenzar a utilizar el modelo, k-Nearest Neighbor debemos encontrar el numero que nos permitira dar con la mejor precision.

```python
from sklearn import metrics
# Para descubrir el k=n que utilizaremos haremos un loop con los numero del 0 al 25. Y Ver cual de estos nos daria el mejor
# resultado.
k_range = range(1, 26)
# Creamos un array llamado datos que nos guardara los datos obtenidos por el modelo para despues graficarlos.
datos = []

# Y ahora usamos el modelo en un for.
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    #Con este comando podemos sacar el porcentaje de exito del modelo con cada k.
    datos.append(metrics.accuracy_score(y_test, y_pred))

print(datos)
```

Como podemos ver tenemos hartos datos, pero vistos de esta forma no queda muy claro, por lo cual pasaremos a graficarlos.

```python
# importamos el pyplot
import matplotlib.pyplot as plt

%matplotlib inline

# Realizamos la relacion entre K y la tasa de exito
plt.plot(k_range, datos)
plt.xlabel('Valores de K por KNN')
plt.ylabel('Tasa de exito')
```

Como podemos ver un valor cerca de 20 nos daria una buena tasa de exito, para corroborar la usaremos en el siguiente codigo.

```python
# Reemplazamos con 19, ya que este es el numero que mayor % nos dio.
n_neighbors = 19
 
knn = KNeighborsClassifier(n_neighbors)
knn.fit(X_train, y_train)
print('Precision del K-NN: {:.2f}'
     .format(knn.score(X_train, y_train)))
```

# Curva AUC - ROC

Midiendo curva para Random Forest (En este caso es mejor y mÃ¡s eficiente con los datos estandarizados)

```python
#Se mide la eficiencia para el modelo de RANDOM FOREST.
from sklearn.metrics import roc_auc_score
#Se obtienen las probabilidades de la prediccion.
prob_RndF = RndF.predict_proba(X_test)
prob_RndF = prob_RndF[:, 1]  
auc = roc_auc_score(y_test, prob_RndF)  
print('AUC: %.2f' % auc) 
```

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, prob_RndF) 
plt.plot(fpr, tpr, color='red', label='ROC')
plt.plot([0, 1], [0, 1], color='blue', linestyle='-')
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Caracteristica Operativa Receptor (Curva ROC)')
plt.legend()
plt.show()
```
